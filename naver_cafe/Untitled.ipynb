```python
driver = webdriver.Chrome(executable_path=f'{cwd}/chromedriver.exe')
driver.get("https://cafe.naver.com/sevenknights")
time.sleep(2)
driver.implicitly_wait(3)
```

```python
search_input = driver.find_element_by_css_selector("#topLayerQueryInput")
search_input.send_keys("레볼루션")
```

```python
push_btn = driver.find_element_by_css_selector("#cafe-search > form > button")
push_btn.click()
driver.switch_to.frame('cafe_main')
```

```python
ass = driver.find_elements_by_xpath("//div[@class='inner_list']/a[@class='article']")
```

```python
print(ass)
```

```python
href_list = []
for a in ass:
    print(a)
    time.sleep(1)
    href_list.append(a.get_attribute('href'))

```

```python
for href in href_list:
    driver.get(href)
    time.sleep(1)
    driver.switch_to_window(driver.window_handles[-1])
    time.sleep(1)
    driver.switch_to.frame("cafe_main")
    time.sleep(1)
    soups = BeautifulSoup(driver.page_source, 'lxml')
    time.sleep(1)
    try:
        title = soups.find("h3",{"class":"title_text"}).text.strip()
        content = soups.find("span",{"class":"se-fs- se-ff- "}).text.strip()
        print(title,content)
    except:
        continue
```

```python
try:
    test_href = href_list[5]
    driver.get(test_href)
    time.sleep(1)
    driver.switch_to_window(driver.window_handles[-1])
    time.sleep(1)
    driver.switch_to.frame("cafe_main")
    time.sleep(1)
    soups = BeautifulSoup(driver.page_source, 'lxml')
except:
    print("error")
    
```

```python
for idx,comment in enumerate(soups.find_all("span",{"class":"text_comment"})):
    comment = ''.join(comment.text.splitlines())
    print(idx,comment)
```
